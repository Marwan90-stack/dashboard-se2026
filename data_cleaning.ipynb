```python
import os
import polars as pl
import plotly.express as px
```

```python
df = pl.read_excel("data/alokasi-petugas.xlsx", sheet_name = "petugas2")
df.head(10)
```

```python
df.columns
```

```python
selected_columns = [
  'nmdesa',
 'nmsls',
 'jenis',
 'nmkec',
 'kk',
 'usaha',
 'muatan',
 'pcl',
 'pml']

selected_columns
```

```python
df2 = df.select(pl.col(selected_columns))
df2.head(10)
```

```python
df2 = (
  df2.with_columns(
  nmdesa = pl.col("nmdesa").str.to_titlecase(),
  nmkec = pl.col('nmkec').str.to_titlecase(),
  pcl = pl.col('pcl').str.to_titlecase(),
  pml = pl.col('pml').str.to_titlecase()
)
.select(
  pl.col(['nmkec', 'nmdesa', 'nmsls', 'jenis', 'pml', 'pcl', 'kk', 'usaha', 'muatan'])
)
)
df2.head(5)
```

```python
df2.write_excel('data/alokasi-petugas-cleaned.xlsx')
```

```python
df2 = (
  df2.with_columns(
    (pl.col("muatan")*0.4).ceil().alias("muatan_40"),
    (pl.col("usaha")*0.4).ceil().alias("usaha_40"),
    (pl.col("usaha")*0.6).ceil().alias("usaha_60"),
    (pl.col("muatan")*0.6).ceil().alias("muatan_60")
  )
  .with_columns(
    (pl.col("muatan_60")/30).ceil().alias("target_harian_ruta_termin1"),
    (pl.col("usaha_40")/30).ceil().alias("target_harian_usaha_termin1"),
    (pl.col("muatan_60")/45).ceil().alias("target_harian_ruta_termin2"),
    (pl.col("usaha_60")/45).ceil().alias("target_harian_usaha_termin2")
  )
  
)

df2.head(10)
```

```python
df2.columns
```

```python
df2 = (
  df2.select(
    pl.col('pml', 'pcl', 'target_harian_ruta_termin1', 'target_harian_usaha_termin1', 'target_harian_ruta_termin2', 'target_harian_usaha_termin2')
  )
)

df2.head(10)
```

```python
target_ppl = (
  df2
  .select(pl.col('pcl', 'target_harian_ruta_termin1', 'target_harian_ruta_termin2', 'target_harian_usaha_termin1', 'target_harian_usaha_termin2'))
  .group_by(pl.col("pcl"))
  .sum()
)

target_ppl
```

```python
target_pml = (
  df2
  .select(pl.col('pml', 'target_harian_ruta_termin1', 'target_harian_ruta_termin2', 'target_harian_usaha_termin1', 'target_harian_usaha_termin2'))
  .group_by(pl.col("pml"))
  .sum()
)

target_pml
```
